# Generating training data

```python
import data_utils
            
reader = data_utils.SpeechCorpusReader('data')

i_sizes = list()
o_sizes = list()
ratios = list()

for i, o in reader.generate_samples('train', 128, infinite=False):
  i_size = i.shape[0]
  o_size = len(o)
  ratio = i_size / o_size

  i_sizes.append(i_size)
  o_sizes.append(o_size)
  ratios.append(ratio)
```

```python
plt.hist(i_sizes, bins=100)
plt.show()
```

```python
plt.hist(o_sizes, bins=100)
plt.show()
```

```python
plt.hist(ratios, bins=100)
plt.show()
```

```python
ax = plt.scatter(i_sizes, o_sizes, s=0.01).axes
ax.set_xlabel('Input size')
ax.set_ylabel('Output size')
plt.scatter([375, 570, 650, 770], [180, 250, 300, 350], color='red', s=1)
plt.show()
```
